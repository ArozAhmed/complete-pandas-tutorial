```python
bios_v3 = bios.copy()
```

```python
bios_v3['born_date'] = pd.to_datetime(bios_v3['born_date'])
```

```python
bios_v3['year_born'] = bios_v3['born_date'].dt.year
bios_v3['month_born'] = bios_v3['born_date'].dt.month
```

```python
bios_v3_final = bios_v3.groupby(['year_born', 'month_born'])['name'].count().reset_index().rename(columns={'name': 'count'}).sort_values(by='count', ascending=0)

bios_v3_final.to_csv('./data/bios_v3_final')
```
